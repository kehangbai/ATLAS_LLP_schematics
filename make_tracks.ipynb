```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_line(a, b, xmax, xmin):
    x = np.linspace( xmax, xmin, 50 ) 
    y = a*x + b
    return x, y
```

```python
def get_circle(cx, cy, r, theta1, theta2):
    angle = np.linspace( theta1, theta2, 50 ) 
    x = cx + r * np.cos( angle ) 
    y = cy + r * np.sin( angle ) 
    return x, y
```

```python
clayer = "dimgrey"
lslayer = (5, (10, 3))
layer_kwargs = {"fill": False, "ls": lslayer, "lw": 2, "color":clayer}
layer2_kwargs = {"fill": False, "ls": "-", "lw": 2, "color":clayer}
layer_pix_kwargs = {"fill": False, "ls": lslayer, "lw": 2, "color":clayer, "alpha":0.5}

cprompt = "lightgrey"
prompt_kwargs = {"ls": "-", "lw": 4, "color":cprompt,"alpha":0.8}

cdis = "deepskyblue"
dis_kwargs = {"ls": "-", "lw": 4, "color":cdis}
dis_d0_kwargs = {"ls": "--", "lw": 4, "color":cdis, "alpha":0.4}

clrt = "limegreen"
lrt_kwargs = {"ls": "-", "lw": 4, "color":clrt}
lrt_d0_kwargs = {"ls": "--", "lw": 4, "color":clrt, "alpha":0.4}

dis2_kwargs = {"ls": "-", "lw": 3, "color":cdis}

lrt2_kwargs = {"ls": "-", "lw": 3, "color":clrt}

ccharge = "mediumpurple"
charge_kwargs = {"ls": "-", "lw": 4, "color":ccharge,"alpha":0.8}
hit_kwargs = {"marker": "o", "s":180, "c":ccharge}

cneutral = "gold"
neutral_kwargs = {"ls": "--", "lw": 4, "color":cneutral}
```

```python
figure, ax = plt.subplots(figsize=(8, 8))
IP = plt.Circle((0,0), 0.3, fill=True, color=clayer)
L0 = plt.Circle((0,0), 6, **layer_kwargs)
L1 = plt.Circle((0,0), 9, **layer_kwargs)
L2 = plt.Circle((0,0), 12, **layer_kwargs)

# Prompt track

x1,y1 = get_circle(cx=-16, cy=0, r=16, theta1=0, theta2=0.28*np.pi)
ax.plot(x1, y1, **prompt_kwargs)

x2,y2 = get_circle(cx=16, cy=0, r=16, theta1=0.72*np.pi, theta2=np.pi)
ax.plot(x2, y2, label='prompt track', **prompt_kwargs)

# Add primary particle
ax.axhline(y=0, xmin=0.33, xmax=0.5, color='gold', linestyle=':', lw=4, label='primary particle 1')
ax.axhline(y=0, xmin=0.5, xmax=0.65, color='pink', linestyle=':', lw=4, label='primary particle 2')

# Displaced track

x3,y3 = get_circle(cx=-5, cy=-12, r=12, theta1=0.5*np.pi, theta2=0.73*np.pi)
ax.plot(x3, y3, label=r'displaced small $d_0$ track', **dis_kwargs)

x4,y4 = get_circle(cx=-5, cy=12, r=12, theta1=-0.5*np.pi, theta2=-0.73*np.pi)
ax.plot(x4, y4, **dis_kwargs)

x3,y3 = get_circle(cx=-5, cy=-12, r=12, theta1=0*np.pi, theta2=0.5*np.pi)
ax.plot(x3, y3, **dis_d0_kwargs)

# LDR track

x5,y5 = get_circle(cx=13, cy=2, r=9, theta1=1.075*np.pi, theta2=1.45*np.pi)
ax.plot(x5, y5, label=r'displaced large $d_0$ track', **lrt_kwargs)

x5,y5 = get_circle(cx=13, cy=-2, r=9, theta1=0.55*np.pi, theta2=0.925*np.pi)
ax.plot(x5, y5, **lrt_kwargs)

x5,y5 = get_circle(cx=13, cy=2, r=9, theta1=0.65*np.pi, theta2=1.08*np.pi)
ax.plot(x5, y5, **lrt_d0_kwargs)

ax.set_aspect( 1 )
ax.add_artist( IP )
ax.add_artist( L0 )
ax.add_artist( L1 )
ax.add_artist( L2 )

# Set axis limits to ensure the circle is visible
ax.set_xlim(-15,15)
ax.set_ylim(-15,15)

plt.title( 'Tracks' )
plt.legend(frameon=True, framealpha=1.0, facecolor='white', edgecolor='none', fontsize=17)
plt.show
plt.savefig("./tracks.png")
```
